```python
# Calculate the average marks for 10th grade, 12th grade, degree, and MBA
average_marks = {
    'ssc_p_avg': placement_data['ssc_p'].mean(),
    'hsc_p_avg': placement_data['hsc_p'].mean(),
    'degree_p_avg': placement_data['degree_p'].mean(),
    'mba_p_avg': placement_data['mba_p'].mean()
}

# Define a function to determine if a student has above average marks in all tests
def above_avg(row, averages):
    return (row['ssc_p'] > averages['ssc_p_avg']) and \
           (row['hsc_p'] > averages['hsc_p_avg']) and \
           (row['degree_p'] > averages['degree_p_avg']) and \
           (row['mba_p'] > averages['mba_p_avg'])

# Apply the function to each row in the dataframe to create the new column
placement_data['above_avg_marks'] = placement_data.apply(above_avg, axis=1, averages=average_marks)

# Convert the boolean column to 'Yes' and 'No'
placement_data['above_avg_marks'] = placement_data['above_avg_marks'].apply(lambda x: 'Yes' if x else 'No')

# Display the updated dataframe with the new column
placement_data[['ssc_p', 'hsc_p', 'degree_p', 'mba_p', 'above_avg_marks']].head()
```
